```
import nltk
alice = nltk.corpus.gutenberg.words('carroll-alice.txt')
# moby = nltk.corpus.gutenberg.words('')
```

```
alice_fd = nltk.FreqDist(alice)
# alice_fd

# This will return the frequency distribution of all the words present inside the document 
```

```
# alice_fd_100 = [words[0] for words in alice_fd]
alice_fd_100 = alice_fd.most_common(100)
# alice_fd_100 --> further producing the most common 100 words present on the text
```

```
moby = nltk.corpus.gutenberg.words('melville-moby_dick.txt')
```

```
moby_fd = nltk.FreqDist(moby)
```

```
moby_fd_100 = moby_fd.most_common(100)
```

```
alice_100 = [words[0] for words in alice_fd_100]
# alice_100

moby_100 = [words[0] for words in moby_fd_100]
# moby_100 
```

```
# alice_100 - moby_100  --------> unsupported operand type(s) for -: 'list' and 'list'
```

```
most_common_on_alice100 = set(alice_100) - set(moby_100)

# The values are present inside the dictionary,so mathematical operation 
# could be possible




# The below results shows that, inside alice_100, set(alice_100) - set(moby_100)
# could give most common words present in inside the alice in wonderland book 

most_common_on_moby100 = set(moby_100)-set(alice_100)
# The vice versa results 


# This could also help to provide some descriptive words when 
# compared to indescriptive words


# This could help in TF-IDF
```

Bigrams concepts 

```
import nltk 

text = 'I think it might rain today.'
tokens = nltk.word_tokenize(text)
tokens
```

```
biagrams = nltk.bigrams(tokens)
for items in biagrams:
  print(items)
```

```
triagrams = nltk.trigrams(tokens)
for items in triagrams:
  print(items)
```

Using ngrams concepts

```
from nltk.util import ngrams
text2 = 'If it is nice out, I will go to the beach'
tokens=nltk.word_tokenize(text2)
bigrams = nltk.bigrams(tokens)
for items in bigrams:
  print(items)
```

```
bigrams = nltk.ngrams(tokens,2)
for items in bigrams:
  print(items)
```

```
fourgrams = nltk.ngrams(tokens,4)
for items in fourgrams:
  print(items)
```

Using the function to write the ngrams

```
def n_grams(text,n):
  tokens = nltk.word_tokenize(text)
  grams = nltk.ngrams(tokens,n)
  return grams
```

```
text3 = text + " " + text2
text3 
```

```
grams = n_grams(text3,5)
for items in grams:
  print(items)
```

Regular Expressions

```
import nltk
import re # Regular Expression
```

```
alice = nltk.corpus.gutenberg.words('carroll-alice.txt')
```

```
set([word for word in alice if re.search("^new",word)])
```

```
[word for word in alice if re.search("^new",word)]

# Understanding the above code without the set() function 
```

```
set([word for word in alice if re.search('ful$',word)])

# In order to return the words that end with -ful, we had used only $ to denote 
# the end of the string  
```

```
# When we using the above code without the set()

[word for word in alice if re.search('ful$',word)]

# From the result, we can found that, certain words are repeated many number of 
#times which is nothing but the duplicates
# So, set() could be able to remove the duplicates and provide results back 
# in dictionary 
```

Various regular expression findings 

```
set([word for word in alice if re.search('^..nn..$',word)])
```

```
set([word for word in alice if re.search('^[chr]at$',word)])
```

```
set([word for word in alice if re.search('^.*nn.*$',word)])

# .* indicates the occurance of 0 or wildcard character for n times 
```

```
len(set([word for word in alice if re.search('^.*nn.*$',word)]))
```

```
set([word for word in alice if re.search('^.*nn.+$',word)])

# includes n wildchart characters in the usage of .+ --> wildcard character to be used atleast once 
# .* same as the above 
```

```
len(set([word for word in alice if re.search('^.*nn.+$',word)]))
```

```
set([word for word in alice if re.search('^[aeiou]c.+y$',word)])
```
